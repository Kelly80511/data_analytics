# Task 1: Environment set-up

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import randn
import matplotlib.pyplot as plot
%matplotlib inline
from numpy.random import randn
```

# Task 2: Data collection

```python
#collecte data
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv',)
df
```

# Task 3: Data wrangling and EDA (exploratory data analysis)

```python
#filtering the data to the most recent 30 days
df30 = df[df['date']>='2020-12-15']
df30
```

```python
#filtering NY as chosen state and NYC as chosen county.
dfnyc = df30[df30['county'] == 'New York City']
dfnyc
```

```python
#dropping the NaN column
drop_cols=['fips']
dfnyc.drop(drop_cols, inplace = True, axis=1)
dfnyc
```

```python
#adding new columns: new cases and new deaths
dfnyc['new cases'] = dfnyc["cases"] - dfnyc["cases"].shift()
dfnyc['new deaths'] = dfnyc['deaths'] - dfnyc['deaths'].shift()
dfnyc
```

# Task 4: Understand NY COVID 19 data in last 30 days

```python
#filtering NY as state
dfny = df30[df30['state'] == 'New York']
dfny
```

```python
#creating a bar plot with counties and cases in New York
sns.barplot(x='county',y='cases',data=dfny)
```

```python
#creating a bar plot with counties and deaths in New York
sns.barplot(x='county',y='deaths',data=dfny)
```

```python
total_cases = dfnyc['cases'].sum()
total_deaths = dfnyc['deaths'].sum()
total_new_cases = dfnyc['new cases'].sum()
total_new_deaths = dfnyc['new deaths'].sum()
print('The total number of cases in New York City is ' + str(total_cases))
print('The total number of deaths in New York City is ' + str(total_deaths))
print('The total number of new cases in New York City is ' + str(total_new_cases))
print('The total number of new deaths in New York City is ' + str(total_new_deaths))
```

```python
sns.pairplot(dfnyc)
```

```python
#creating a bar plot with date and new cases in NYC
sns.barplot(x='date',y='new cases',data=dfnyc)
```

```python
#creating a bar plot with date and new deaths in NYC
sns.barplot(x='date',y='new deaths',data=dfnyc)
```

According to the COVID 19 data of New York State in the current 30 days, New York City has the most COVID 19 cases in New York State, including cases, deaths, new cases, and new deaths.

# Task 5: Comparing COVID 19 data from different states

```python
#creating a bar plot with state and total number of cases
df30.groupby(['state'])['cases'].sum().plot(kind='bar')
```

```python
#creating a pie plot with state and total number of cases
df30.groupby(['state'])['cases'].sum().plot(kind='pie')
```

```python
#the states with the most cases in turn is: California, Texas, Florida...
df30.groupby(['state'])['cases'].sum()
```

```python
#creating a bar plot with state and total number of deaths
df30.groupby(['state'])['deaths'].sum().plot(kind='bar')
```

```python
#creating a pie plot with state and total number of deaths
df30.groupby(['state'])['deaths'].sum().plot(kind='pie')
```

```python
#the states with the most deaths in turn is: New York, Texas, California...
df30.groupby(['state'])['deaths'].sum()
```

According to the COVID 19 data of the U.S. in the current 30 days, the states with the most cases, in turn, are California, Texas, Florida...  The states with the most deaths, in turn, are New York, Texas, California...
